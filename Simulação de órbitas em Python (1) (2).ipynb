```python
from math import *

#conjcorpos = ['Terra', 'Marte', 'Urano']
#corpo = 'Terra'
def aceleracao_corpo(conjcorpos, corpo):
    global G 
    G = 6.67384E-11

    #recolhe massas corpos em contribuicao
    massascorpos = []
    for nome in conjcorpos:
        massa = df.loc[df['nome'] == nome, 'massa'].iloc[0]
        massascorpos.append(massa)
    
    #recolhe coordenadas corpos em contribuicao - tem que ser buscadas as coords atualizadas em cada iteracao....
    coordenadascorpos = []
    for nome in conjcorpos:
        c1 = (df.loc[df['nome'] == nome, 'lx'].iloc[0], df.loc[df['nome'] == nome, 'ly'].iloc[0], df.loc[df['nome'] == nome, 'lz'].iloc[0])
        coordenadascorpos.append(c1)
        
    #recolhe coordenadas corpo que sofre aceleracao - tem que ser buscadas as coords atualizadas em cada iteracao...
    coordenadascorpo = [df.loc[df['nome'] == corpo, 'lx'].iloc[0], df.loc[df['nome'] == corpo, 'ly'].iloc[0], df.loc[df['nome'] == corpo, 'lz'].iloc[0]]
    
    #cria lista e aplica metodo de euler-cromer
    aceleracoes = []
    for i in range(len(massascorpos)):
        if corpo != conjcorpos[i]: #p cada corpo em contribuição
            ac = ((G*massascorpos[i])/((((coordenadascorpo[0]-coordenadascorpos[i][0])**2)+((coordenadascorpo[1]-coordenadascorpos[i][1])**2)+((coordenadascorpo[2]-coordenadascorpos[i][2])**2))**(3/2))*(coordenadascorpos[i][0]-coordenadascorpo[0]), (G*massascorpos[i])/((((coordenadascorpo[0]-coordenadascorpos[i][0])**2)+((coordenadascorpo[1]-coordenadascorpos[i][1])**2)+((coordenadascorpo[2]-coordenadascorpos[i][2])**2))**(3/2))*(coordenadascorpos[i][1]-coordenadascorpo[1]), (G*massascorpos[i])/((((coordenadascorpo[0]-coordenadascorpos[i][0])**2)+((coordenadascorpo[1]-coordenadascorpos[i][1])**2)+((coordenadascorpo[2]-coordenadascorpos[i][2])**2))**(3/2))*(coordenadascorpos[i][2]-coordenadascorpo[2]))
            aceleracoes.append(ac)   
    
    #cria tuplo c/ componentes da aceleracao p cada corpo
    componentesac = sum(i[0] for i in aceleracoes), sum(i[1] for i in aceleracoes), sum(i[2] for i in aceleracoes)
    
    return componentesac

#aceleracao_corpo(conjcorpos, corpo)
```

Começou-se por recolher as massas dos corpos presentes na lista do conjunto de corpos para os quais a órbita vai ser desenhada, recorrendo ao método loc que vai buscar os valores em questão ao dataframe. Este método é executado para todos os corpos através de um ciclo for, e as suas massas são posteriormente adicionadas a uma lista, assegurando-se a sequencialidade das massas na lista em relação aos corpos presentes na lista "conjcorpos".

Segue-se a mesma lógica para cada uma das coordenadas (em x, y e z) para cada um dos corpos, que são armazenadas sob a forma de tuplo dentro de uma lista (permitindo um fácil acesso corpo/coordenada). A escolha da utilização de tuplos como estrutura de recolha das coordenadas dos corpos em contribuição prende-se com o facto de atribuir uma forma de leitura tangível e percetível relativamente ao tipo de dado que aquele tuplo alberga (no caso, componentes de coordenadas 3D), sendo mais intuitiva a busca de cada uma das componentes na forma de tuplo do que lista.

Posteriormente, executa-se o método de Euler-Cromer (assegurando que este só é corrido entre o corpo e o corpo do conjunto de corpos em questão quando estes forem diferentes um do outro, evitando-se divisões por zero e a presença de "nan" nas componentes da aceleração a ser somadas), e faz-se a soma das acelerações em cada uma das componentes, colocando cada uma delas na variável componentesac, também na forma de tuplo (pela mesma razão anteriormente descrita).

```python
from math import *

#conjcorpos = ['Terra', 'Marte', 'Urano']
#deltat = 20
def velocidade_apos_dt(conjcorpos, deltat):
    
    velocidadescorpos = []
    for elem in conjcorpos:
        v = (df.loc[df['nome'] == elem, 'vx'].iloc[0], df.loc[df['nome'] == elem, 'vy'].iloc[0], df.loc[df['nome'] == elem, 'vz'].iloc[0])
        velocidadescorpos.append(v)
    
    aceleracoescorpos = []
    for elem in conjcorpos:
        ac = aceleracao_corpo(conjcorpos, elem)
        aceleracoescorpos.append(ac)
    
    velatualiz = []
    for i in range(len(velocidadescorpos)):
        vel = (velocidadescorpos[i][0] + ((aceleracoescorpos[i][0])*deltat), velocidadescorpos[i][1] + ((aceleracoescorpos[i][1])*deltat), velocidadescorpos[i][2] + ((aceleracoescorpos[i][2])*deltat))
        velatualiz.append(vel)
    
    for elem in conjcorpos:
        for i in range(len(conjcorpos)):
            df.loc[df['vx'] == velocidadescorpos[i][0], 'vx'] = velatualiz[i][0]
            df.loc[df['vy'] == velocidadescorpos[i][1], 'vy'] = velatualiz[i][1]
            df.loc[df['vz'] == velocidadescorpos[i][2], 'vz'] = velatualiz[i][2]
    
    return velatualiz

#falta implementar os dicionarios que guardam a localizacao atual (? ou so na seguinte, em q guardo simult. pos e loc?)
#velocidade_apos_dt(conjcorpos, deltat)
```
